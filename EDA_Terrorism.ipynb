```python
# Check for missing values
df_dropna = df_drop.dropna(axis=0)
df_dropna
```

```python
# Drop the duplicates
df_clean = df_dropna.drop_duplicates().reset_index(drop=True).copy()
df_clean
```

```python
# Rename the columns
df_clean = df_clean.rename(columns={
    'iyear': 'Year', 'imonth': 'Month', 'iday': 'Day', 'country_txt': 'Country', 'region_txt': 'Region', 'provstate': 'Province/State', 
    'city': 'City', 'attacktype1_txt': 'Attack_Type', 'targtype1_txt': 'Target_Type', 'corp1': 'Company', 'target1': 'Target',
    'natlty1_txt': 'Nationality', 'gname': 'Group_Name', 'weaptype1_txt': 'Weapon_Type', 'nkill': 'Number_Killed', 'nwound': 'Wounded',
    'nhostkid': 'Kid_Hostage_Number', 'kidhijcountry': 'Kid_Nationality',
    'ransomamt': 'Ransom_Amount', 'nreleased': 'Released_Hostage_Number', 'dbsource': 'Data_Source'})
df_clean
```

## Step 4: Data Visualization

```python
# Find the terrorism hot zone (Indicate total and average incidents per city)
ax = df_clean.query("City != 'Unknown'") \
            .groupby('City')['City'] \
            .agg(['count']) \
            .query('count >= 10') \
            .plot(kind='bar', figsize=(10,5), title="Cities with Highest Terrorism Incidents")

ax.set_xlabel("City")
ax.set_ylabel("Count")
plt.savefig("TerrorismHotspots.png")
plt.show()
```

```python
# Check which nationalities are common targets
ax = df_clean.query("Nationality != 'Unknown'") \
                .groupby('Nationality')['Nationality'] \
                .agg(['count']) \
                .query('count >= 10') \
                .plot(kind='bar', figsize=(10,5), title="Countries Targeted by Terrorists")

ax.set_xlabel("Nationality")
ax.set_ylabel("Count")
plt.savefig("CountriesTargeted.png")
plt.show()
```

```python
# Find the Weapon Type used for Terrorism Activities
ax = df_clean.query('Weapon_Type != "Unknown"') \
                .groupby('Weapon_Type')['Weapon_Type'] \
                .agg(['count']) \
                .query('count >= 5') \
                .plot(kind='bar', figsize=(10,5), title="Weapon Types Used by Terrorists")

ax.set_xlabel("Weapon")
ax.set_ylabel("Count")
plt.savefig("TerroristWeaponType.png")
plt.show()
```
